```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



# path of the CSV file for each quater 
file_Path_Q1 = "C:/Users/berzaf/Desktop/DS4A/data/quarterly location data/Financials_L1_formatted_Q1_20200331.csv"
file_Path_Q2 = "C:/Users/berzaf/Desktop/DS4A/data/quarterly location data/Financials_L1_formatted_Q2_20200630.csv"
file_Path_Q3 = "C:/Users/berzaf/Desktop/DS4A/data/quarterly location data/Financials_L1_formatted_Q3_20200930.csv"
file_Path_Q4 = "C:/Users/berzaf/Desktop/DS4A/data/quarterly location data/Financials_L1_formatted_Q4_20201231.csv"
#file_Path_ppp = "C:/Users/berzaf/Desktop/DS4A/data/quarterly location data/public_up_to_150k_1_230331.csv"
file_Path_ppp = "C:/Users/berzaf/Desktop/DS4A/data/quarterly location data/public_150k_plus_230331.csv"



# Load the CSV file into a pandas DataFrame
df_1 = pd.read_csv(file_Path_Q1)
df_2 = pd.read_csv(file_Path_Q2)
df_3 = pd.read_csv(file_Path_Q3)
df_4 = pd.read_csv(file_Path_Q4)
df_ppp = pd.read_csv(file_Path_ppp)




# Display the DataFrame
#df_1.head()
df_ppp.head()

#print(df_1)
```

```python
df_combined = pd.concat([df_1, df_2, df_3, df_4])
```

```python
# Grouping data by "State" and calculating the average "Total Assets" and average deposits for each state
average_assets_by_state = df_combined.groupby('State')['Total Assets'].mean()
average_deposits_by_state = df_combined.groupby('State')['Total Deposits'].mean()
```

```python
print(average_assets_by_state,average_deposits_by_state)
```

```python
#plot average assets by state
plt.figure(figsize=(12, 6))  

plt.bar(average_assets_by_state.index, average_assets_by_state)
plt.xlabel('State')
plt.ylabel('Average Total Assets')
plt.title('Average Total Assets by State')

plt.xticks(rotation=90)  # Rotate the x-labels 

plt.show()
```

```python

# Average Total Assets and Total Deposits by state
plt.figure(figsize=(12, 6)) 

# Bar width
bar_width = 0.35
bar_positions = np.arange(len(average_assets_by_state))

# Plotting Average Total Assets
plt.bar(bar_positions, average_assets_by_state, bar_width, label='Average Total Assets')

# Plotting Total Deposits
plt.bar(bar_positions + bar_width, average_deposits_by_state, bar_width, label='Total Deposits')

plt.xlabel('State')
plt.ylabel('Amount (in billions)')
plt.title('Average Total Assets and Total Deposits by State')
plt.xticks(bar_positions + bar_width/2, average_assets_by_state.index, rotation=90)  # Set x-axis ticks and labels

plt.legend()  # Display legend

plt.show()
```

```python
# Sort the DataFrame by total deposits in descending order
sorted_df = df_combined.sort_values(by="Total Deposits", ascending=False)

# Select the top N banks
top_n = 30
top_banks = sorted_df.head(top_n)

# Plot the bar chart
#plt.figure(figsize=(12, 6))
plt.figure(figsize=(8, 8))


plt.bar(top_banks["Institution Name"], top_banks["Total Deposits"])
plt.xlabel("Institution Name")
plt.ylabel("Total Deposits")
plt.title(f"Total Deposits - Top {top_n} Institutions")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python


# Create a list of quarter names
quarters = ["Q1", "Q2", "Q3", "Q4"]

# Get the total deposits for each quarter
deposits = [
    df_1["Total Deposits"].sum(),
    df_2["Total Deposits"].sum(),
    df_3["Total Deposits"].sum(),
    df_4["Total Deposits"].sum()
]


# Plot the total deposits for each quarter
plt.figure(figsize=(10, 8))
plt.plot(quarters, deposits, marker='o')

plt.xlabel("Quarter")
plt.ylabel("Total Deposits")
plt.title("Total Deposits by Quarter")
plt.grid(True)
plt.show()
```

```python
# get the bank names

lender_names = df_ppp["OriginatingLender"].tolist()

# Filter the DataFrames to include only the matching bank names
df_1_filtered = df_1[df_1["Institution Name"].isin(lender_names)]
df_2_filtered = df_2[df_2["Institution Name"].isin(lender_names)]
df_3_filtered = df_3[df_3["Institution Name"].isin(lender_names)]
df_4_filtered = df_4[df_4["Institution Name"].isin(lender_names)]

# Create a list of quarter names
quarters = ["Q1", "Q2", "Q3","Q4"]

# Get the total deposits for each bank and quarter
bank_deposits = []

for df, quarter in zip([df_1_filtered, df_2_filtered, df_3_filtered,df_4_filtered], quarters):
    bank_deposits.append((quarter, df["Institution Name"], df["Total Deposits"]))

# Plot the total deposits for each bank and quarter
plt.figure(figsize=(10, 6))

for i, (quarter, lender_names, deposits) in enumerate(bank_deposits):
    plt.plot(lender_names, deposits, marker='o', linestyle='-', label=quarter)

plt.xlabel("Institution Name")
plt.ylabel("Total Deposits")
plt.title("Total Deposits by Bank and Quarter")
plt.xticks(rotation=90)

# Add a legend with handles and labels
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels)

plt.grid(True)
plt.show()
```
